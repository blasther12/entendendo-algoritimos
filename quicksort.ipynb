## Quicksort



```python
import random

def quicksort(array):
    if len(array) < 2:
        return array  # Caso Base: arrays com 0 ou 1 elemento já estão ordenados
    else:
        # Caso recursivo ** Nem sempre escolher o primeiro elemento como o pivô vai ser a melhor opção.
        # Buscar sempre um pivô aleatório
        pivo = random.choice(array)

        menores = [i for i in array if i < pivo]   # Subarray de todos os elementos menores que o pivô
        iguais  = [i for i in array if i == pivo]  # Elementos iguais ao pivô (importante quando há repetidos)
        maiores = [i for i in array if i > pivo]   # Subarray de todos os elementos maiores que o pivô

        return quicksort(menores) + iguais + quicksort(maiores)
```

```python
print(quicksort([10, 5 , 2, 3, 7, 8, 6, 10, 23, 5, 10]))
```

```python
def quicksort(array):
    if len(array) < 2:
        return array  # Caso Base: arrays com 0 ou 1 elemento já estão ordenados
    else:
        # Caso recursivo ** Nem sempre escolher o primeiro elemento como o pivô vai ser a melhor opção.
        # Aqui usamos um pivô no meio da lista para evitar casos ruins de forma simples.
        pivo = array[len(array) // 2]

        menores = [i for i in array if i < pivo]   # Subarray de todos os elementos menores que o pivô
        iguais  = [i for i in array if i == pivo]  # Elementos iguais ao pivô (evita perder valores repetidos)
        maiores = [i for i in array if i > pivo]   # Subarray de todos os elementos maiores que o pivô

        return quicksort(menores) + iguais + quicksort(maiores)
```

```python
print(quicksort([10, 5 , 2, 3, 7, 8, 6, 10, 23, 5, 10]))
```
