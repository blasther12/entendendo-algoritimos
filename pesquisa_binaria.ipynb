## **Notação Big O**

A notação de Big O é uma notação especial que diz o quão rapido é um algoritimo.

*A notação Big O estabelece o tempo de execução para a pior hipótese *

- O(log n) -> Tempo logarítimo. Exemplo: Pesquisa binaria.
- O(n) -> Tempo linear. Exemplo: pesquisa simples.
- O(n * log n) -> Exemplo: um algoritimo rápido de ordenação, como a ordenação quicksort.
- O(n²) -> Exemplo: Um algoritimo lento de ordenação, como a ordenação por seleção.
- O(n!) -> Um algoritimo bastante lento, como o do caixeiro viajante.

A pesquisa binária é um algoritmo que recebe uma lista ordenada. Com a lista ordenada, fazemos um chute, que é escolher um valor no meio da lista, e verificamos se esse item é maior ou menor do que o que estamos procurando. Assim, reduzimos o intervalo de busca a cada passo.

```python
minha_lista = [1, 3, 5, 7, 9]

def pesquisa_binaria(lista, item):
  baixo = 0 # baixo e alto acompanham a parte da lista que estamos procurando
  alto = len(lista) - 1

  while baixo <= alto: # Enquanto ainda não conseguiu chegar a um unico elemento
    meio = (baixo + alto) // 2 # Verifica o elemento central
    chute = lista[meio]
    if chute == item: # achou o item
      return meio
    if chute > item: # Chute foi muito alto
      alto = meio - 1
    else: # Chute muito baixo
      baixo = meio + 1
  return None
```

```python
print(pesquisa_binaria(minha_lista, 3))
print(pesquisa_binaria(minha_lista, -1))
```
